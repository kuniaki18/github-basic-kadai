```python
# 台形の寸法
top_side = 10   # 上辺（cm）
bottom_side = 20  # 下辺（cm）
height = 5      # 高さ（cm）

# 台形の面積 =（上辺＋下辺）× 高さ ÷ 2
area = (top_side + bottom_side) * height / 2

# 結果を出力
print(f"{area}cm²")
```
